```python
# buscar os dados de Operaçoes.csv,
import pandas as pds
import numpy

path_file = "../assets/Operaçoes.csv"

df = pds.read_csv(path_file, encoding='UTF_8', sep=",")

display(df)
```

```python
# modificar as datas, e remover colunas desnecessárias"

df['Data'] = pds.to_datetime(df['Data'])

df_extrato_operacoes = df.drop(columns=["Origem", "Situação"])

display(df_extrato_operacoes)
```

```python
# Filtrar somente os dados de 2021

start_date = "2021-01-01"
end_date = "2021-12-31"

date_filter_template = (df_extrato_operacoes["Data"] >= start_date) & (df_extrato_operacoes["Data"] <= end_date)

df_extrato_operacoes = df_extrato_operacoes.loc[date_filter_template]

display(df_extrato_operacoes)
```

```python
df_extrato_operacoes['Preço'] = df_extrato_operacoes['Preço'].str.replace(',', '.')

# converter a coluna Preço(object) para Preço(float)
df_extrato_operacoes['Preço'] = pds.to_numeric(df_extrato_operacoes['Preço'], errors='coerce')

# Obter o valor total da operação
df_extrato_operacoes['Custo'] = df_extrato_operacoes['Quantidade'].multiply(df_extrato_operacoes['Preço'])

for index in df_extrato_operacoes.index:
          if df_extrato_operacoes['Operação'][index] == 'Venda':
                    df_extrato_operacoes['Custo'][index].values() * 10
#           else:
#                     df_extrato_operacoes['Custo'] = df_extrato_operacoes['Custo'][index]

# df_extrato_operacoes['Operação'].values[[11]]


display(df_extrato_operacoes)
```
